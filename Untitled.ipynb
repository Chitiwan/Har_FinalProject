```python
def windows(data, size):
    start = 0
    while start < data:
        yield int(start), int(start + size)
        start += (size / 2)
x =[]
yt=[]

f = 1
for i in alldata1:
    for (start, end) in windows(len(i["ACC_X"]), 396):
        ax = i["ACC_X"][start:end]
        ay = i["ACC_Y"][start:end]
        az = i["ACC_Z"][start:end]
        x.append([np.mean(ax),np.mean(ay),np.mean(az),np.std(ax),np.std(ay),np.std(az)])
        yt.append([1,0,0]) #hot-one
        
    f +=1
f = 1
for i in alldata3:
    for (start, end) in windows(len(i["ACC_X"]), 396):
        ax = i["ACC_X"][start:end]
        ay = i["ACC_Y"][start:end]
        az = i["ACC_Z"][start:end]
        x.append([np.mean(ax),np.mean(ay),np.mean(az),np.std(ax),np.std(ay),np.std(az)])
        yt.append([0,1,0]) #hot-one
    f +=1
f = 1
for i in alldata4:
    for (start, end) in windows(len(i["ACC_X"]), 396):
        ax = i["ACC_X"][start:end]
        ay = i["ACC_Y"][start:end]
        az = i["ACC_Z"][start:end]
        x.append([np.mean(ax),np.mean(ay),np.mean(az),np.std(ax),np.std(ay),np.std(az)])
        yt.append([0,0,1]) #hot-one
    f +=1
    
```

```python
x=np.array(x)
yt=np.array(yt)
```

```python
def g(x):    
    return 1/(1+np.exp(-x))

def grad_g(g):
        return (1-g)*g



#random init weight and bias
np.random.seed(1)
N0=6 #first layer 4 nodes
N1=3 #second layer 2 nodes
eta=0.01
a0=np.ones([x.shape[0],x.shape[1]+1])
a0[:,:-1]=x


w0 = np.random.random((6,N0)) # 6 inputs x 4 nodes
b0 = np.random.random((1,N0)) # 1 bias x 4 nodes
wb0=np.concatenate((w0, b0), axis=0)

a1=np.ones([x.shape[0],N0+1])
w1 = np.random.random((N0,N1)) # 4 inputs x 2 nodes
b1 = np.random.random((1,N1)) # 1 bias x 2 nodes
wb1=np.concatenate((w1, b1), axis=0)

for i in range(200000):
    #forward x=a0, a2=yp
    z0=np.dot(a0,wb0)
    a1[:,:-1]=g(z0)
    #a1=np.concatenate([g(z0),np.ones([a0.shape[0],1])], axis=1)
    
    z1=np.dot(a1,wb1)
    a2=g(z1)
    
    #backward
    d_a2=yt-a2
    d_z1=d_a2*grad_g(a2)
    wb1 += eta*np.dot(a1.T,d_z1)
    
    d_a1=np.dot(d_z1,wb1.T)
    d_z0=d_a1*grad_g(a1)
    wb0 += eta*np.dot(a0.T,d_z0[:,:-1])
    
    if(i % 10000) == 0:   # Only print the error every 10000 steps
        E=0.5*np.sum(np.square(d_a2))
        print("Error: {}".format(E))
```
