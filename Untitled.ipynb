```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
plt.style.use('ggplot')

def feature_normalize(dataset):
    mu = np.mean(dataset,axis = 0)
    sigma = np.std(dataset,axis = 0)
    return (dataset - mu)/sigma

def read_data(file_path):
    data = pd.read_csv(file_path)
    return data


alldata1 = []
alldata3 = []
alldata4 = []
for i in range(1,16):
    for file in os.listdir("./Rowdata66Hz/SubjectNo."+str(i)+"/neck/activity"):
        if file.split("_")[2] in ["1"]:
            alldata1.append(read_data("./Rowdata66Hz/SubjectNo."+str(i)+"/neck/activity/"+file))
        elif file.split("_")[2] in ["3"]:
            alldata3.append(read_data("./Rowdata66Hz/SubjectNo."+str(i)+"/neck/activity/"+file))
        elif file.split("_")[2] in ["4"]:
            alldata4.append(read_data("./Rowdata66Hz/SubjectNo."+str(i)+"/neck/activity/"+file))



```
